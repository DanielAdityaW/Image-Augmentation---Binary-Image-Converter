```python
import numpy
import glob
import os
import cv2
import matplotlib.pyplot as plt
from PIL import Image
```

```python
directorya = r"D:\Documents\Deep Learning Project\Handsign\new_handsign_images\Images\Test\A"
os.mkdir('A_')

i = 0
for img in glob.glob(directorya + "/*.jpg"):
    new_image = cv2.imread(img, cv2.IMREAD_GRAYSCALE)
    #color = cv2.applyColorMap(image, cv2.COLORMAP_WINTER)
    _, image = cv2.threshold(new_image, 115, 255, cv2.THRESH_BINARY_INV) 
    
    cv2.imwrite("A_/A%04i.jpg" %i, image)
    i += 1
    cv2.imshow('image', image)
    cv2.waitKey(30)

cv2.destroyAllWindows()


"""cv2.imshow('gambar', image)

cv2.waitKey(0)
cv2.destroyAllWindows()

cv2.imwrite('gambar.jpg', image)"""
```
